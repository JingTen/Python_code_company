```python
# -*- coding: <encoding name> -*- : # -*- coding: utf-8 -*-
'''
车型库 匹配燃油&工信部组合之外数据（燃油进口）
'''

import csv
import ast
import codecs

# 定义判断函数
def StrListInStr(str_list,str2):
    for i in str_list:
        if i not in str2:
            return False
    return True

def utf8_2_gbk(oldfile):
    # 设置新文件名称
    newfile = "gbk_%s" % oldfile

    f = codecs.open(oldfile, 'r', 'utf-8')
    utf_str = f.read()
    f.close()

    out_gbk_str = utf_str.encode('GB18030')

    f = open(newfile, 'wb')
    f.write(out_gbk_str)
    f.close()

# 燃油自定义编号，车系车型名称读取
# 初始化燃油车系字典
rycx_dict = {}

with open('rycx.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        k = line[0]
        v = line[1]
        rycx_dict[k] = v
csvfile.close()

# 车型库车系ID与车型名称读取
# 初始化车型库车系字典
cxkcx_dict = {}

with open('cxkcx.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        k = line[0]
        v = line[1]
        cxkcx_dict[k] = v
csvfile.close()


# 初始化结果列表
result = []
# 遍历燃油车系字典
for k1,v1 in rycx_dict.items():
    rycx_bh = k1
    rycx_name_list = v1.split(' ')
    
    print('%s---正在匹配...' % rycx_bh)
    
    # 遍历车型库车系字典
    for k2,v2 in cxkcx_dict.items():
        cxk_name = k2
        cxk_seriesId = v2
        
        # 判断燃油车系名称是否在车型库名称中出现
        if StrListInStr(rycx_name_list,cxk_name):
            result.append([rycx_bh,cxk_seriesId])
            
# 结果写入CSV文件
resultfile = open('rycx_result.csv','w',newline='',encoding='utf-8')

writer = csv.writer(resultfile)

for i in result:
    writer.writerow(i)

resultfile.close()

# 结果文件转换格式
utf8_2_gbk("rycx_result.csv")

print('结果输出完毕！')           
            

```
