```python
'''
Excel

111ABC
111, BDC
222, EDE
333, LFAAF
333, FAALD
...

csv


'''

import csv
import ast
import codecs


list_all = []

with open('.csv','r') as csvfile:
    reader = csv.reader(csvfile)

    for line in reader:
        
        list_all.append(ast.literal_eval(line[0]))

csvfile.close()


dic = {}
for i in list_all:
    dic.setdefault(i[0],[]).append(i[1])
    
print(dic)
def utf8_2_gbk(oldfile):
    # 
    newfile = "gbk_%s" % oldfile

    f = codecs.open(oldfile, 'r', 'utf-8')
    utf_str = f.read()
    f.close()

    out_gbk_str = utf_str.encode('GB18030')

    f = open(newfile, 'wb')
    f.write(out_gbk_str)
    f.close()
    
# CSV
resultfile = open('_result.csv','w',newline='',encoding='utf-8')

writer = csv.writer(resultfile)

for k,v in dic.items():
    writer.writerow([k,v])

resultfile.close()

# 
utf8_2_gbk('_result.csv')

print('')

```
