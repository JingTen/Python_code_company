```python
'''
将Excel类似

111，ABC
111, BDC
222, EDE
333, LFAAF
333, FAALD
...

数据以每行两个元素组成的一个列表csv文件
转化为一对多形式的汇总

'''

import csv
import ast
import codecs


list_all = []

with open('力洋整车型号修复.csv','r') as csvfile:
    reader = csv.reader(csvfile)

    for line in reader:
        
        list_all.append(ast.literal_eval(line[0]))

csvfile.close()


dic = {}
for i in list_all:
    dic.setdefault(i[0],[]).append(i[1])
    
print(dic)
def utf8_2_gbk(oldfile):
    # 设置新文件名称
    newfile = "gbk_%s" % oldfile

    f = codecs.open(oldfile, 'r', 'utf-8')
    utf_str = f.read()
    f.close()

    out_gbk_str = utf_str.encode('GB18030')

    f = open(newfile, 'wb')
    f.write(out_gbk_str)
    f.close()
    
# 结果写入CSV文件
resultfile = open('力洋整车型号修复_result.csv','w',newline='',encoding='utf-8')

writer = csv.writer(resultfile)

for k,v in dic.items():
    writer.writerow([k,v])

resultfile.close()

# 结果文件转换格式
utf8_2_gbk('力洋整车型号修复_result.csv')

print('结果输出完毕！')

```
