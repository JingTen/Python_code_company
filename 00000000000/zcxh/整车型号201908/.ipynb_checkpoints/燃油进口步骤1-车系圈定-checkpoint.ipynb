```python
# -*- coding: <encoding name> -*- : # -*- coding: utf-8 -*-
'''
车型库 匹配燃油&工信部组合之外数据（燃油进口）
'''

import csv
import ast
import codecs

# 定义判断函数
def StrListInStr(str_list,str2):
    for i in str_list:
        if i not in str2:
            return False
    return True


# 燃油自定义编号，车系车型名称读取
# 初始化燃油车系字典
rycx_dict = {}

with open('rycx.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        k = line[0]
        v = line[1]
        rycx_dict[k] = v
csvfile.close()

# 车型库车系ID与车型名称读取
# 初始化车型库车系字典
cxkcx_dict = {}

with open('cxkcx.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        k = line[0]
        v = line[1]
        cxkcx_dict[k] = v
csvfile.close()


# 初始化结果列表
result = []
# 遍历燃油车系字典
for k1,v1 in rycx_dict.items():
    rycx_bh = k1
    rycx_name_list = v1.split(' ')
    
    # 遍历车型库车系字典
    for k2,v2 in cxkcx_dict.items():
        cxk_name = k2
        cxk_seriesId = v2
        
        # 判断燃油车系名称是否在车型库名称中出现
        if StrListInStr(rycx_name_list,cxk_name):
            result.append([rycx_bh,cxk_seriesId])
```
