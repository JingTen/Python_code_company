```python
# coding:utf-8



import csv

def read_csv_file(file_name):
    '''读取csv文件，结果为以每行数据为列表（元素为每列属性）的总列表'''
    try:
        with open(file_name,'r') as file:
            contents = file.readlines()
    except FileNotFoundError:
        print('文件%s不存在' % file_name)
        return None
    else:
        f6 = [] 
        for line in contents:
            f6.append(line.strip().split(','))

    return f6

def model_split(model):
    '''分解F6的model为单词组成的列表'''
    if model.find('[') != -1 and model.find(']') != -1:
        model_1 = model[:model.find('[')].strip()
        model_2 = model[model.find('[')+1:model.find(']')].strip()

        model_1_list = model_1.split()
        model_2_list = model_2.split()
        model_list = [model_1_list,model_2_list]
    else:
        model_list = [model.split()]
    
    return model_list
'''
# 处理文件名称
file_name_f6 = 'F6.csv'
file_name_f6_che = 'F6_CHE.csv'


# 读取文件
f6 = read_csv_file(file_name_f6)
f6_che = read_csv_file(file_name_f6_che)
'''
# F6结果
f6_result = []

# 车型库结果
che_result = {}

# 测试 
'力帆汽车力帆汽车乐途 S1.5手动变速箱(MT)2015'
'力帆汽车力帆汽车乐途 S1.5手动变速箱(MT)2015'
f6 = [['ad2009cd8cf56dd9d8a15d4617a5b503','力帆汽车力帆汽车乐途 S1.5手动变速箱(MT)2015','乐途 S']]
f6_che = [['1003663','力帆汽车力帆汽车乐途 S1.5手动变速箱(MT)2015','乐途 2015款 1.5L S 基本型']]

# 遍历F6
count_print = 0
for i in f6:
    count_print += 1
    print('第%s条记录正在匹配' % count_print)
    
    f6_id = i[0]
    f6_condition = i[1]
    f6_model = i[2]
    
    # 遍历车型库
    for j in f6_che:
        che_id = j[0]
        che_condition = j[1]
        che_model = j[2]
        
        # 先判断条件是否一致，一致则继续，否则开始下一个车型库的车型
        if f6_condition == che_condition:
            # 再判断model所有的描述性单词是否包含在车型库的车型名称中
            
            # 处理F6中的model:
            f6_model_lists = model_split(f6_model)
        
            for f6_model_list in f6_model_lists:
                # 计数
                count = 0
                for k in f6_model_list:
                    # 遍历每个单词，符合条件（即，在车型名称中）则计数+1
                    if k in che_model:
                        count += 1
                # 当计数为 model_condition 的长度时，表示所有描述性单词都在车型名称中，符合条件
                if count == len(f6_model_list):
                    # 加入F6结果
                    f6_result.append([f6_id,che_id])
                    # 加入车型库结果
                    if che_id in che_result:
                        che_result[che_id].append(f6_id)
                    else:
                        che_result[che_id] = [f6_id]
print('车型匹配处理完毕')
```

```python
import time
import codecs

def utf8_2_gbk(oldfile):
    # 设置新文件名称
    newfile = "gbk_%s" % oldfile

    f = codecs.open(oldfile, 'r', 'utf-8')
    utf_str = f.read()
    f.close()

    out_gbk_str = utf_str.encode('GB18030')

    f = open(newfile,'wb')
    f.write(out_gbk_str)
    f.close()

def f6_write_csvfile(lists):
    with open('f6_result.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(lists)

def che_write_csvfile(dicts):
    with open('che_result.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        for k,v in dicts.items():
            writer.writerow([k,str(v)])
            
# 结果写入csv文件
f6_write_csvfile(f6_result)
che_write_csvfile(che_result)

time.sleep(5)
# 转码

utf8_2_gbk('f6_result.csv')
utf8_2_gbk('che_result.csv')
print('结果写入处理完毕')
```
