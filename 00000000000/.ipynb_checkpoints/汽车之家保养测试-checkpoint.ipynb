```python
import requests
from bs4 import BeautifulSoup


# 请求头设置
session = requests.Session()

model_id = 16200

url_body = "https://car.autohome.com.cn/mtn/spec/cycle/"
url = url_body + str(model_id)



headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.6 Safari/537.36",
    #"Content-Type": "application/x-www-form-urlencoded",
    #"Referer": url,
    #"Origin": "https://car.autohome.com.cn"
}

html = session.get(url,headers=headers).content
# html2 = requests.get(url, headers=headers).content


bsObj = BeautifulSoup(html, 'lxml')

print(bsObj)
```

```python
# 车辆标准保养项目 及 合计费用
maintenance_program = ['发动机机油','机油滤清器','变速箱油','后刹车片','火花塞','空调滤清器','空气滤清器','防冻液','前刹车片','燃油滤清器','正时皮带','制动液','转向助力油','工时费（总计）','配件费（总计）','合计']

# 表格提取
table = bsObj.find('table',{'id':'tableFixed'})

thead = table.find('thead')
tbody = table.find('tbody').find_all('tr')
tfoot = table.find('tfoot')

# 各个车型ID下结果总列表
id_result = []

# 保养项目_公里数列表
byxm_km = []
thead_text = thead.find_all('td')
for i in thead_text:
    byxm_km.append(i.get_text().strip().replace('\n','/'))

# 初始化配件费字典、工时费字典
parts_price_dict = {
    '发动机机油': [],
    '机油滤清器': [],
    '变速箱油': [],
    '后刹车片': [],
    '火花塞': [],
    '空调滤清器': [],
    '空气滤清器': [],
    '防冻液': [],
    '前刹车片': [],
    '燃油滤清器': [],
    '正时皮带': [],
    '制动液': [],
    '转向助力油':[]
}
man_hours_price_dict = {
    '发动机机油': [],
    '机油滤清器': [],
    '变速箱油': [],
    '后刹车片': [],
    '火花塞': [],
    '空调滤清器': [],
    '空气滤清器': [],
    '防冻液': [],
    '前刹车片': [],
    '燃油滤清器': [],
    '正时皮带': [],
    '制动液': [],
    '转向助力油':[]
}

# 保养项目_具体项目列表
for i in range(len(tbody)):
    # 当循环至工时费，配件费总计（最后两条记录）记录时，另行处理
    if i < len(tbody) - 2:
        # 初始化单项标准保养项目列表
        byxm_single = []
        byxm_single_data = tbody[i].find_all('td')
        for j in byxm_single_data:
            # 存在a标签，则为单项标准保养项目具体名称（如发动机机油，机油滤清器...）
            if j.find('div',{'class':'column-tit'}):
                byxm_name = j.find('div',{'class':'column-tit'}).get_text()
                byxm_single.append(byxm_name)
            # 不存在a标签，则存在p标签，为单项标准保养项目在某公里数下是否需要保养(通常，第一个p标签表示是否需要保养，第二个p标签为配件费，第三个p标签为工时费)
            else:
                all_p = j.find_all('p')
                print(all_p)
                # 第一个p标签-是否需要保养字段加入单个保养项目
                byxm_single.append(all_p[0].get_text().replace('\xa0',''))
                # 若有3个p标签，则包含工时费和配件费，若只有一个p标签，则无费用信息
                if len(all_p) == 3:
                    parts_price = all_p[1].get_text().replace('配件费','')
                    man_hours_price = all_p[2].get_text().replace('工时费','')
                    # print(parts_price,man_hours_price)
                    if byxm_name in parts_price_dict:
                        parts_price_dict[byxm_name].append(parts_price)
                        man_hours_price_dict[byxm_name].append(man_hours_price)
        #print(byxm_single)
        id_result.append(byxm_single)
    else:
        # 初始化工时费或配件费列表
        byxm_price = []
        byxm_price_data = tbody[i].find_all('td')
        for k in byxm_price_data:
            byxm_price.append(k.get_text().replace('¥','').strip())
        #print(byxm_price)
        id_result.append(byxm_price)

# 保养项目_合计列表
all_price = []
all_price_data = tfoot.find_all('td')
for p in all_price_data:
    all_price.append(p.get_text().replace('¥','').strip())
id_result.append(all_price)


# 各个车型ID保养项目结果存入该车型ID字典
id_dict = {}
# 首先字典中加入所有公里数项目，并初始化其所有的具体项目值为空
for num in range(1,len(byxm_km)):
    id_dict[byxm_km[num]] = {'发动机机油': '',
                            '机油滤清器': '',
                            '变速箱油': '',
                            '后刹车片': '',
                            '火花塞': '',
                            '空调滤清器': '',
                            '空气滤清器': '',
                            '防冻液': '',
                            '前刹车片': '',
                            '燃油滤清器': '',
                            '正时皮带': '',
                            '制动液': '',
                            '转向助力油': '',
                            '工时费（总计）': '',
                            '配件费（总计）': '',
                            '合计': ''}
# 然后遍历车型ID结果总列表，每个具体保养项目加入字典中对应的公里数项目小字典中
for i in id_result:
    if i[0] in maintenance_program:
        for num in range(1,len(i)):
            id_dict[byxm_km[num]][i[0]] = i[num]

```
