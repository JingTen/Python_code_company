```python
import pymysql
from collections import Counter


conn = pymysql.connect(host='127.0.0.1', user='root', passwd='123456', db='MySQL', charset='utf8')
cur = conn.cursor()
cur.execute("USE gm_chevrolet")

sql = '''
    SELECT optionCodes FROM `gm_chevrolet`.`vin_hash` 
    WHERE `model` = 'C140' AND `series` = 'LP' AND `body` = '26' AND LEFT(vin,8) = 'LSGLP83X'
'''

cur.execute(sql)
results = cur.fetchall()

# 获取所有不重复配置
all_codes = []
for i in results:
    all_codes.extend(eval(i[0]))
    
# 每个配置代码出现的次数以字典形式存储
all_codes_count = Counter(all_codes)
print(all_codes_count['GFT'])

# 列表去重
all_codes = list(set(all_codes))

# 对原始列表进行排序
all_codes.sort()

# 转化为tuple，方便SQL语句导入变量
all_codes = tuple(all_codes)

# 查询所有code对应的中文描述
sql = '''
    SELECT optionCode,descriptionCN FROM vin_codes
    WHERE optionCode in {}
    ORDER BY optionCode;
'''.format(all_codes)

cur.execute(sql)
results = cur.fetchall()
conn.close()

for i in results:
    print(i)

```

```python
import pymysql
from collections import Counter


conn = pymysql.connect(host='127.0.0.1', user='root', passwd='123456', db='MySQL', charset='utf8')
cur = conn.cursor()
cur.execute("USE gm_chevrolet")

sql = '''
    DELETE FROM `vin_result`
    WHERE
        `vin_result`.Model = 'C140'
        AND `vin_result`.Series = 'LP'
        AND `vin_result`.Body = '26'
        AND `vin_result`.VIN8 = 'LSGLP84X';
'''

cur.execute(sql)
results = cur.fetchall()

print(results)
```
