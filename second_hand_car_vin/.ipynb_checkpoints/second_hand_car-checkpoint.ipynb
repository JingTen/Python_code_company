```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time
import requests

# 设置请求头
session = requests.Session()

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/49.0.2623.112 Safari/537.36",
          "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"}


# 第三步，获取每辆二手车HTML中的车型id,vincode
def get_html(url):
    html = session.get(url,headers=headers)

    bsObj = BeautifulSoup(html.text.'lxml')
    
    #print(bsObj)

    script = bsObj.find('body').find('script',{'type':'text/javascript'}).get_text()

    re_id = re.compile(r"specid: '\d*'")
    re_vincode = re.compile(r"vincode: '\w*'")
    
    # 确保网页中未找到车型id和vincode时，程序正常运行，并返回空列表
    try:
        result_id = re.search(re_id, script).group()
        result_vincode = re.search(re_vincode, script).group()

        id_vincode_list = []
        id_vincode_list.append(url)
        id_vincode_list.append(result_id)
        id_vincode_list.append(result_vincode)
    except:
        id_vincode_list = []

    return id_vincode_list

# 判断每辆二手车的url地址类型并格式化为手机端url
def judge_and_format_url(url):
    # 手机端url前部
    mobile_url_head = "https://m.che168.com"
    
    if "https" in url:
        new_url = mobile_url_head + long_url_change(url)
    elif "/dealer/" or "/lianmeng/" in url:
        new_url = mobile_url_head + url
    else:
        pass
    
    return new_url

# html为长链接时，访问并提取两个Id，方便生成所需网址
def long_url_change(url):
    html = session.get(url,headers=headers)

    bsObj = BeautifulSoup(html.text.'lxml')

    mobile_agent = str(bsObj.find('head').find('meta',{'http-equiv':'mobile-agent'}))

    re_url = re.compile(r"/(dealer|lianmeng)/\d*/\d*\.html")

    result_url = re.search(re_url, mobile_agent).group()
    
    return result_url







html = "https://semnt.autohome.com.cn/monitor?rtype=2&dpmk=9OFAl0QfSefFMg%2FR0tF%2F4Jow0%2FuzciafhE3M7eA8fsfXIpKn9TB%2Fw9krkbLgdCqDiHaDpOA%2FepDYJMSr8Ct%2Bx9YmjKqjNSzfgCfE%2Bu8%2Be5fRJJf4o2R%2FxoMwjbyycCaWwyiDp%2FdkdsSEJ5Oz8zN4xswmwK2hKy7HhHaMpvZjf8TXdMD88jItrdIllKjgKm2ThWLO7atibcjDIZar9CRj0JRhxOyrYm3IwzCNvLJqLoaHfdPz4Dxtw9EjkK7wNn%2FC0DCNvLdoJoaIdoOk8TZ5x9ghjbyhdCqTlXvX%2B6tibcjSIZeo%2BjN83sNx1eeyY23I0z6D%2F6FlJpbDKJar9DR23sNx1O2rYm3I1SOTqfE3Y9CCc8z3piR1wdEllqf7Km2GhH%2FI%2BuA8fsTYPoPusGk5m49xxLz4JH7C0yORqfs%2Ff8LRIoOy4GUmhpgwm7zzNn3D0SWYp%2Fs%2Ff8LDPoPzo34QgpN7wvvgPHnCzTDD96ZZP4CIccS8%2BDN7ws0wwvGxchCfjnbEvPg3Y9CCfdLqnWAukZV907z4NmHD0CKYpvswf8LVI5in%2Bz5%2Bxc0wwvKrZSStlGDNvPgkJ4aVYtKk7Sk4hZY8wvanN3nKz3HO8%2B1iKpONd9Ox9DR%2FytI9kqz0NnrC1SeP9rZrI82RZMDsp2cmltwjkav3MH3QzTDC%2F7ZjKJ2Ta4Ok8TB%2Bw80wxvGtYjytiHaDpPA%2BecPWJY28sWk6gIJ3g6TwKm2bkk3O6qpjPZGIZti8%2BDdj0JFzxvudZSaGmDCbvPEyf8PRIoPj"
#html = "/dealer/113985/31366397.html?pvareaid=100519&userpid=340000&usercid=340100"
#html = "/dealer/304818/32149809.html?pvareaid=100519&userpid=340000&usercid=340100"

"https://www.che168.com/guangzhou/bieke/a0_0msdgscncgpiltocsp1exv100x0q0/"


format_url = judge_and_format_url(html)

result = get_html(format_url)

print(result)

```
