```python
# coding=utf-8
# 解析投诉问题类型
# 初始化投诉问题类型

import requests
from bs4 import BeautifulSoup

url = 'http://www.12365auto.com/js/cTypeInfo.js?version=20191120'
headers = {
    'Referer': 'http://www.12365auto.com/zlts/0-0-0-0-0-0_0-0-1.shtml',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.6 Safari/537.36'
}

r = requests.get(url,headers=headers,timeout=6)
response = eval(r.text.replace('var cTypeInfo = ','').replace('\r\n',''))

ts_question_types = []
for i in response:
    big_id = i['id']
    name = i['name']
    value = i['value']
    zf = i['zf']
    if zf == 'z':
        zf_name = '质量问题'
    elif zf == 'f':
        zf_name = '服务问题'
    else:
        zf_name = ''
    items = i['items']
    
    for ii in items:
        small_id = ii['id']
        title = ii['title']
        code = value + str(small_id)
        # print(big_id,name,value,zf,zf_name,small_id,title,code)
        ts_question_types.append([big_id,name,value,zf,zf_name,small_id,title,code])
print(ts_question_types)



```

```python
# coding=utf-8
# 解析车质网车系

import requests
from bs4 import BeautifulSoup

url = 'http://www.12365auto.com/js/brandsHaveSeries.js?version=20191119'
headers = {
    'Referer': 'http://www.12365auto.com/zlts/0-0-0-0-0-0_0-0-1.shtml',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.6 Safari/537.36'
}

r = requests.get(url,headers=headers,timeout=6)
r = eval(r.text.replace('var brandsHaveSeries = ','').replace(';\r\n','').replace('\n',''))
series_lists = []
for i in r:
    id = i['id']
    name = i['name']
    initials = i['initials']
    brand_logo = i['brand_logo']
    brand_logo = brand_logo.replace('http://img.12365auto.com/f','')
    brand_logo = 'http://img.12365auto.com/f' + brand_logo
    big_config = i['config']
    for ii in big_config:
        brandsId = ii['brandsId']
        brandsName = ii['brandsName']
        brandsIsInlet = ii['brandsIsInlet']
        small_config = ii['config']
        for iii in small_config:
            seriesId = iii['seriesId']
            seriesName = iii['seriesName']
            seriesAttribute = iii['seriesAttribute']
            series_logo = iii ['series_logo']
            series_logo = series_logo.replace('http://img.12365auto.com/f','')
            series_logo = 'http://img.12365auto.com/f' + series_logo
            series_lists.append([id,name,initials,brand_logo,brandsId,brandsName,brandsIsInlet,seriesId,seriesName,seriesAttribute,series_logo])
print(len(series_lists))
print(series_lists)
```

```python
# coding=utf-8
# 解析投诉详细内容
import requests
from bs4 import BeautifulSoup
import time

# ------获取投诉简述跳转链接的详细文本------
def get_tsnrANDtshf(url):
    # 初始化投诉内容和投诉回复列表
    tsnrANDtshf_list = []
    # 请求头设置
    session = requests.Session()
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.6 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3"}

    html = session.get(url, headers=headers, timeout=10)
    if html.status_code == 200:
        bsObj = BeautifulSoup(html.content, 'lxml')
        # 投诉内容有图片或者视频时，其文本获取方式与投诉回复不一致
        tsnr = bsObj.find('div', {'class': 'tsnr'}).find_all('p')[-1].get_text()
        tsnr = bsObj.find('div', {'class': 'tsnr'}).find_all('p')
        tsnr_deal = ''
        for i in tsnr:
            if i.get_text():
                tsnr_deal += i.get_text()

        tshf = bsObj.find('div', {'class': 'tshf'}).find('p').get_text()
        tsnrANDtshf_list.append(tsnr_deal)
        tsnrANDtshf_list.append(tshf)
    elif html.status_code == 404:
        tsnrANDtshf_list = ['网页不存在','网页不存在']
    else:
        print('%s 响应错误' % url)
    
    html.close()
    return tsnrANDtshf_list

url = 'http://www.12365auto.com/zlts/20190121/342107.shtml'

r = get_tsnrANDtshf(url)
print(r)

```

```python
# coding=utf-8
# 解析投诉页面列表
import requests
from bs4 import BeautifulSoup
import time

page = 1
url = 'http://www.12365auto.com/zlts/0-0-0-0-0-0_0-0-{}.shtml'.format(page)
headers = {
    'Referer': 'http://www.12365auto.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.6 Safari/537.36'
}
r = requests.get(url,headers=headers,timeout=10)
if r.status_code == 200:
    bsObj = BeautifulSoup(r.content,'lxml')
    table = bsObj.find('div',{'class': 'tslb_b'}).find('table')
    print(table)
```

```python
tr = table.find_all('tr')[1:]
ts_page_list = []
# 遍历每条投诉信息
for i in tr:
    single_ts = []
    info = i.find_all('td')
    # (1).获取投诉编号，投诉品牌，投诉车系，投诉车型，投诉简述，典型问题，投诉时间，投诉状态 8 项目基本信息
    for j in info:
        single_ts.append(j.get_text())
    # (2).获取品牌id，车系id，车型id，服务问题
    bid = info[2].attrs['bid']
    sid = info[2].attrs['sid']
    mid = info[2].attrs['mid']
    fw = info[2].attrs['fw']
    # (3).获取每条投诉详细内容url
    description_url = i.find('td',{'class': 'tsjs'}).find('a').attrs['href']
    for k in [bid,sid,mid,fw,description_url]:
        single_ts.append(k)
    ts_page_list.append(single_ts)
for k in ts_page_list:
    print(k)
    print('\n')
    
    
```
